```python
# Check for duplicate rows
duplicates = train_df.duplicated().sum()

print(f"Number of duplicate rows: {duplicates}")
```

```python
##Removing Stopwords
import nltk
import string
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re
def clean_text(text):
    text = text.strip() ##Removing leading and trailing spaces
    text = re.sub(r'\d+', '', text) ##Removing numbers
    text = re.sub(r'[^\w\s]', '', text)  # Removing special characters
    text = text.translate(str.maketrans('', '', string.punctuation)) ##Removing punctuation
    text = text.lower()  ##Converting text to lowercase
    stop_words = set(stopwords.words('english'))
    tokens = word_tokenize(text)  # Tokenizing text
    tokens = [word for word in tokens if word not in stop_words]  # Removing stopwords
    lemmatizer = WordNetLemmatizer()  # Lemmatizing text
    tokens = [lemmatizer.lemmatize(word) for word in tokens]
    return ' '.join(tokens)

def clean_dataframe(df):
    for column in df.columns:
        df[column] = df[column].astype(str).apply(clean_text)
    return df
train_cleaned = clean_dataframe(train_df)
test_cleaned = clean_dataframe(test_df)
```

## **Modeling**
<a href=#cont>Back to Table of Contents</a>

```python
X_train = train_cleaned['content']
X_test = test_cleaned['content']
y_train = train_cleaned['category']
y_test = test_cleaned['category']
```

```python
##Label Encoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y_train= le.fit_transform(y_train)
y_test= le.fit_transform(y_test)
```

```python
##Initialize Vectorizers
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
count_vectorizer = CountVectorizer()
tfidf_vectorizer = TfidfVectorizer()
```

```python
# Fit and transform training data with vectorizers
X_train = tfidf_vectorizer.fit_transform(X_train)
X_test = tfidf_vectorizer.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
##Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
RFC_classifier= RandomForestClassifier(random_state=0, n_estimators =100, criterion ='entropy')
model= RFC_classifier.fit(X_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, precision_score,recall_score
from sklearn.metrics import classification_report, accuracy_score
y_pred = model.predict(X_test)
acc = accuracy_score(y_test, y_pred)
weighted_precision = precision_score(y_test, y_pred, average='weighted')
weighted_recall = recall_score(y_test, y_pred, average='weighted')
weighted_f1 = f1_score(y_test, y_pred, average='weighted')
print(classification_report(y_test, y_pred))
model_results = pd.DataFrame([['Random Forest Classifier', acc, weighted_precision, weighted_recall, weighted_f1]],
               columns = ['Model', 'Accuracy', 'Weighted-averaged precision', 'Weighted-averaged recall', 'Weighted-averaged f1_score'])
results= model_results
model_results
```

```python
cm = confusion_matrix(y_test, y_pred)

# Plot the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='g', cmap='Blues', xticklabels=np.unique(y_test), yticklabels=np.unique(y_test))
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix for Multiclass Classification')
plt.show()
```

```python
##Support Vector Machine
from sklearn.svm import SVC
svm_model = SVC(kernel='linear')
svm_model.fit(X_train, y_train)
```

```python
y_predSVC = svm_model.predict(X_test)
acc = accuracy_score(y_test, y_predSVC)
weighted_precision = precision_score(y_test, y_predSVC, average='weighted')
weighted_recall = recall_score(y_test, y_predSVC, average='weighted')
weighted_f1 = f1_score(y_test, y_predSVC, average='weighted')
print(classification_report(y_test, y_predSVC))
model_results = pd.DataFrame([['SVC(linear)', acc, weighted_precision, weighted_recall, weighted_f1]],
               columns = ['Model','Accuracy', 'Weighted-averaged precision', 'Weighted-averaged recall', 'Weighted-averaged f1_score'])
results = results.append(model_results, ignore_index = True)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn_model = KNeighborsClassifier(n_neighbors=3)
knn_model.fit(X_train, y_train)
```

```python
y_predKNN = knn_model.predict(X_test)
acc = accuracy_score(y_test, y_predKNN)
weighted_precision = precision_score(y_test, y_predKNN, average='weighted')
weighted_recall = recall_score(y_test, y_predKNN, average='weighted')
weighted_f1 = f1_score(y_test, y_predKNN, average='weighted')
print(classification_report(y_test, y_predKNN))
model_results = pd.DataFrame([['KNN', acc, weighted_precision, weighted_recall, weighted_f1]],
               columns = ['Model','Accuracy', 'Weighted-averaged precision', 'Weighted-averaged recall', 'Weighted-averaged f1_score'])
results = results.append(model_results, ignore_index = True)
```
